#**DATASET LOADING**

```python
from sklearn.datasets import load_breast_cancer
```

```python
Data = load_breast_cancer()

df = pd.DataFrame(Data.data,columns=Data.feature_names)
```

#**TRAIN TEST SPLIT :**

```python
X_train,x_test,Y_train,y_test = train_test_split(x,y,test_size=0.2,random_state=2)
```

#**DECISION TREE MODEL :**

```python
Doctor = DecisionTreeClassifier(random_state=42)
Doctor.fit(X_train,Y_train)
```

#**CREATE PREDICTION :**

```python
y_pred = Doctor.predict(x_test)
```

```python

Accuracy = Doctor.score(x_test,y_test)
Accuracy
```

```python
from enum import CONFORM
CFM=confusion_matrix(y_test,y_pred)
print("placement confusion matrix:",CFM)
```

```python
CFM = confusion_matrix(y_test, y_pred)
```

#**CLASSIFICATION REPORTS :**

```python
CFR = classification_report(y_test,y_pred)
print(CFR)
```

#**VISUALIZATION USING HEATMAPS :**

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
Labels = Data.target_names

plt.figure(figsize=(8,6))

sns.heatmap(CFM,annot=True,fmt="d",xticklabels=Labels,yticklabels=Labels,cmap="Reds")

plt.title("CONFUSION MATRIX - HEATMAP")
plt.xlabel("Predicted Condition")
plt.ylabel("Actual Condition")

plt.show()
```
